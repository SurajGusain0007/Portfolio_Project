### Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

#### 1.1Loading and reading the dataset

```python
zomato_data=pd.read_csv("zomato.csv")
zomato_data.head()
```

#### 1.2Looking at the information about the dataset,datatypes of the corresponding columns and missing values

```python
zomato_data.info()
```

#### 1.3  Dropping the unnecessary the columns
#### Dropping the column "dish_liked","phone","url" and saving the new dataset as"zomato"

```python
zomato=zomato_data.drop(["url","phone","dish_liked"],axis=1)
```

#### 1.4 Removing the duplicates

```python
zomato.duplicated().sum()
zomato.drop_duplicates(inplace=True)
```

#### 1.5 Remove the Nan values from the dataset


```python
zomato.isnull().sum()
zomato.dropna(how="any",inplace=True)
```

#### 1.6 Info() function is used to get a concise summary of the dataset)

```python
zomato.info()
```

#### 1.7 Reading the column names

```python
zomato.columns
```

#### 1.8 Changing the column name

```python
zomato=zomato.rename(columns={'listed_in(type)':'type',"listed_in(city)":'city',"approx_cost(for two people)":"cost"})
zomato.columns
```

#### 1.9 Some transformations


```python
# Change the cost to string
zomato['cost']=zomato['cost'].astype(str)

# using lambda function to replace "," to "."
zomato['cost']=zomato['cost'].apply(lambda x :x.replace(',','.'))

# Change the cost to float

zomato['cost']=zomato['cost'].astype(float)
zomato.info()
```

#### 1.10 Reading rate of dataset

```python
zomato['rate'].unique()
```

#### 1.11 Removing /5 from dataset

```python
zomato=zomato.loc[zomato.rate!="NEW"]
zomato=zomato.loc[zomato.rate!="-"].reset_index(drop=True)
remove_slash=lambda x: x.replace("/5"," ")  if type(x) == np.str else x
zomato.rate=zomato.rate.apply(remove_slash).str.strip().astype('float')
zomato['rate'].head()
```

#### 1.12 Adjust the columns

```python
zomato.online_order.replace(('Yes','No'),(True,False),inplace=True)
zomato.book_table.replace(('Yes','No'),(True,False),inplace=True)
```

#### 1.13 Encode the input variables

```python

def Encode(zomato):
    for column in zomato.columns[~zomato.columns.isin(['rate','cost','votes','online_order','book_table'])]:
        zomato[column]=zomato[column].factorize()[0]
    return zomato
        
zomato_en=Encode(zomato.copy())
```

```python
zomato_en
```

#### 1.14 Corelation between different variables

```python
corr=zomato_en.corr(method="kendall")
plt.figure(figsize=(16,10))
sns.heatmap(corr,annot=True)
plt.show()
```

The highest correlation between name and address which is 0.62 which is not very much concern

# Data Visualization

#### 2.2 Restaurant allowing table or not

#### 2.3 How many restaurants in every city

#### 2.6 Type and Rating

```python

type_plt=pd.crosstab(zomato['rate'],zomato['type'])
type_plt.plot(kind='bar',stacked=True)
plt.title('Type-Rating',fontsize=15,fontweight='bold')
plt.xlabel('Rating',fontsize=15,fontweight='bold')
plt.ylabel('Type',fontsize=15,fontweight='bold')
plt.xticks(fontsize=10,fontweight='bold')
plt.yticks(fontsize=10,fontweight='bold')
plt.show()
```

#### 2.7 Cost of Restaurant

#### 2.8 Number of restaurants in location

```python
plt.figure(figsize=(18,14))
sns.countplot(x='location',data=zomato,palette='Set1').set_xticklabels(sns.countplot(x='location',data=zomato).get_xticklabels(),rotation=90,ha='right')
plt.xlabel('Location',size=18,fontsize=15,fontweight='bold')
plt.ylabel("Frequency",size=18,fontsize=15,fontweight='bold')
plt.title("No of restaurants ina location",size=20,pad=20,fontsize=15,fontweight='bold')
plt.xticks(fontsize=10,fontweight='bold')
plt.yticks(fontsize=10,fontweight='bold')
plt.show()
```

#### 2.9 Most Famous Restaurants in Bengalure
